# Part 1 - Combine Fa20 Course Info with Requirements

## 1. Read the 2020 course schedule into a DataFrame


```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
with open("spring_courses.html") as f:
    dom = BeautifulSoup(f, 'lxml')
```

```python
number = []
name = []
professors = []
meet_time = []
room = []
des = []
d = 0
```

```python
for row in dom.select("ul.schedule-listing > li"):
    for span in row.select('span'):
        
        if span.select('span'):
            d += 1
            break
        t = span.get_text().strip()
        if d%6 == 0:
            if t[:2] == 'CS':
                number.append(t[:13] + t[14:17])
            elif t[:2] == 'MA':
                number.append(t[:14] + t[15:18])
        elif d%6 == 1:
            if t[:7] == "Special":
                x = t.split('\n')
                name.append(x[0] + ' ' + x[-1].strip())
            else:
                name.append(t)
        elif d%6 == 2:
            professors.append(t.split('\n')[0])
        elif d%6 == 3:
            meet_time.append(t)
        elif d%6 == 4:
            room.append(t)
        elif d%6 == 5:
            des.append(t)
        d += 1
```

```python
dom.select("ul.schedule-listing > li")[0].select('span')[0]
```

```python
schedule = pd.DataFrame(data = {'Number-Section': number, 'Name':name, 'Instructor':professors, 'Time':meet_time})

schedule.sample(10)

schedule['Number'] = schedule['Number-Section'].str.extract('(\w+\-\w+\.\d+)\-\d+')
schedule['Section'] = schedule['Number-Section'].str.extract('\w+\-\w+\.\d+\-(\d+)')
```

## 2. Read the 2020 course catalog into a DataFrame

```python
with open("course_catlog.html") as f:
    dom = BeautifulSoup(f, 'lxml')
```

```python
Number = []
Prereqs = []
Points = []
for row in dom.select('ul.courses-listing > li'):
    info = row.select('p')   #list of elements each row
    Number.append(re.findall('\w+\-\w+\.\d+', info[0].get_text().strip())[0])
    Points.append(re.findall('.+?(?=Points)', info[1].get_text())[0].strip())
    Prereqs.append(re.findall('\: (.+)', info[2].get_text())[0])
```

```python
catlog = pd.DataFrame({'Number': Number, 'Prereqs': Prereqs, 'Points': Points})
```

## 3. Put together both DataFrames

```python
pd.set_option('display.max_rows', 200)
f = pd.merge(schedule, catlog, how='left', on='Number')
del(f['Number-Section'])
del(f['Section'])
f.reindex(columns=['Number', 'Name', 'Instructor', 'Time', 'Prereqs', 'Points'])
```

No obvious anomalies. The behavior of how="left" keeps all entries of the first DataFrame, which in this case is schedule, to merge catlog entries into schedule based on the intersection column 'Number'
